```python
import numpy as np
import matplotlib.pyplot as plt

def brownian_motion(num_steps, dt):
    # Generate random increments
    increments = np.sqrt(dt) * np.random.randn(num_steps)

    # Calculate the cumulative sum to get the Brownian motion
    brownian_path = np.cumsum(increments)

    return brownian_path

# Set parameters
num_steps = 1000  # Number of steps
dt = 0.01        # Time step

# Generate Brownian motion
bm_path = brownian_motion(num_steps, dt)

# Plot the Brownian motion
plt.plot(np.arange(0, num_steps*dt, dt), bm_path)
plt.title('Brownian Motion Simulation')
plt.xlabel('Time')
plt.ylabel('Position')
plt.show()
```

```python
# Parameters for the random walk
n_steps_rw = 200
n_simulations_rw = 100000  # Increase number of simulations for more accuracy

# Simulate the random walk
def simulate_random_walk(n_steps):
    # Steps: +1 or -1 with equal probability
    steps = np.random.choice([-1, 1], size=n_steps)
    # Cumulative sum to get the path
    path = np.cumsum(steps)
    return path

# Simulation
final_values = []  # Store final values of S_200 for walks where max S_n >= 10
for _ in range(n_simulations_rw):
    path = simulate_random_walk(n_steps_rw)
    if np.max(path) >= 10:  # Check if the path meets the condition
        final_values.append(path[-1])

# Expected value
expected_value_condition = np.mean(final_values) if final_values else 0
expected_value_condition
```
